```python
import cv2
import os
from matplotlib import pyplot as plt
```

```python
# 2. For all images in data dir
```

```python
# all images in data set: https://isic-archive.com/api/v1/image
# API https://isic-archive.com/api/v1/#/image
dir = os.path.join('data', 'images', 'CanField')
for file in sorted(os.listdir(dir)):
    # 1. Skip non-images
    if not file.endswith(".jpg"):
        continue # next
    print(file)
    
    # 2. Read in Image
    img_path = os.path.join(dir, file)
    img = cv2.imread(img_path)
    
    # 3. Render Image
    plt.imshow(img) # show original image
    plt.show()
    plt.close()
    
    # 4. Edge detection
    
    recolor = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(recolor, cv2.COLOR_BGR2GRAY)
    recolor = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    
    blurred = cv2.GaussianBlur(img, (3,3),0)
    canny = cv2.Canny(blurred, threshold1=180, threshold2=200)
    
    plt.imshow(recolor)      # show next image
    plt.show()
    plt.close()
    
    plt.imshow(blurred)   # show next image
    plt.show()
    plt.close()
    
    plt.imshow(canny)    # show edges
    plt.show()
    plt.close()
    
    # 5. Write out Image
    rename = file.replace(".jpg", "") + "_edge.jpg"
    output_path = os.path.join('data','output', rename)
    print(rename)
    
    cv2.imwrite(output_path, gray)
    
    
```
